```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## reading the data  from the url into a dataframe with the needed headerfiles
```

```python
filename = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
df.head()
```

```python
## now we need to identify the missing values 
```

```python
## replace the '?'  with Nan values 
df.replace('?',np.nan,inplace = True)   ## when we set inplace = True the changes will be reflected in the same dataframe df 

## if we wanted df to remain the same way we should have done df1 = df.replace('?',np.nan, inplace = False)
```

```python
df.head()
```

```python
## Evaluating the missing data 
```

```python
df_missing = df.isnull()
df_missing.head()
```

```python
## visualize missing values in each column 
## value_counts()  - returns a series with unique values and their counts 
```

```python
for col in df_missing.columns.values:
    print(col)
    print(df_missing[col].value_counts())
    print(" ")
```

```python
### now how do we deal with this missing data ? 
## 1. drop whole row 
## 2. drop whole col (only do it if you think the feature is not going to contribute as much and there are more missing values than useful data)
## 3. replace data with mean
## 4. replace with frequency 
## 5. replace based on custom functions
```

```python
## dropping the whole row 
## since the price is the target value we will drop all rows with missing price values

df_clean = df[df['price'].notna()]
```

```python
## another way to remove the rows is with df.dropna
df_c = df.dropna(subset = "price",axis = 0)

## df.dropna(subset = "price",axis = 0,inplace = True)
```

```python
df_c
```

```python
df_clean.head()
```

```python
## check if price has any nan values (sanity check)
sanitydf = df_clean[['price']].isnull()
sanitydf['price'].value_counts()
```

```python
## all the rows with nan values for price  have been removed from the data set 
```

```python
df_clean.head()
```

```python
print(df_clean['num-of-doors'].value_counts())
```

```python
## making a copy
dfc = df_clean.copy()
```

```python
missing_list = ["horsepower","peak-rpm","normalized-losses","stroke","bore"]
for miss in missing_list:
    print('Unique values in ',miss)
    print(df[miss].nunique())
    
```

```python
## all these features have too many unique values to replace with the frequency method
## we can replace them with the mean method
```

```python
## calculate the mean of normalized losses
nl_mean = dfc['normalized-losses'].astype('float').mean(axis = 0)
nl_mean
```

```python
dfc['normalized-losses'].replace(np.nan,nl_mean,inplace = True)
```

```python
## sanity check
sanitydf = dfc['normalized-losses'].isnull()
sanitydf.value_counts()
```

```python
## calculate the mean for the rest of the columns
missing = ["horsepower","peak-rpm","stroke","bore"]
for miss in missing:
    mean = dfc[miss].astype('float').mean(axis = 0)
    dfc[miss].replace(np.nan,mean,inplace = True)
```

```python
## sanity check for all the rows
sanitydf = dfc.isnull()
for col in sanitydf.columns.values:
    print(col)
    print(sanitydf[col].value_counts())
```

```python
## nan values removed/replaced from all the feature
```

```python
## next we will make sure the data has a proper datatype
dfc.head()
```

```python
dfc.dtypes
```

*There are many numerical values that have a data type that is not int or float so we will need to convert them to a usable datatype*

```python
## identify the numerical features 
numlist = ["normalized-losses","bore","stroke","horsepower","peak-rpm","price"]

for num in numlist:
    dfc[[num]] = dfc[[num]].astype('float')
```

## Data Standardization ##

```python
dfs = dfc.copy()
```

Standardization is the process of transforming data into a common format which allows the researcher to make the meaningful comparison.

```python
## convert mpg to L/100km by dividing 235 with the mpg value 
dfs['city-mpg'] = 235/dfs['city-mpg']
dfs['highway-mpg'] = 235/dfs['highway-mpg']
```

```python
dfs.head()
```

**Normalization**<br>

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1

*We can either normalize with by dividing the value with the maximum value or we can normalize using the mean and standard deviation*

```python
dfs['length'] = dfs['length']/dfs['length'].max()
```

```python
dfs['width'] = dfs['length']/dfs['length'].max()
```

```python
dfs['height'] = dfs['height']/dfs['height'].max()
```

**BINNING** <br>

Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

```python
## first plot the histogram for horsepower or the feature you wish to transform
```

```python
%matplotlib inline   
## to make sure the graphs are plotted within the jupyter notebook
```

```python
## histogram 
plt.hist(dfs['horsepower'])
plt.title('Horsepower bins')
plt.xlabel('horsepower')
plt.ylabel('count')
```

```python
## check how many unique values there are
dfs['horsepower'].nunique()
```

```python
## we want to transform the data such that there are only 3 categories - low, medium and high 
## if we use linspace(start = , end = , num = )  - generates equally spaced numbers between the intervals mentioned

## num indicates the number of divisions - if we need 3 equal bins or 3 categories there would be 4 dividers 
```

```python
hp = np.linspace(dfs['horsepower'].min(), dfs['horsepower'].max(), 4)
```

```python
hp
```

```python
group_names = ['low','medium','high']
```

```python
## we can use pd.cut(to_be_binned, num of bins, labels = group names, include_lowest = True) to do the binning 
dfs['horsepower-binned'] = pd.cut(dfs['horsepower'],3, labels = group_names, include_lowest = True)
```

```python
dfs[['horsepower','horsepower-binned']].head(10)
```

```python
## visualising the bins

plt.bar(group_names, dfs['horsepower-binned'].value_counts())
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('Bins')
```

```python
plt.hist(dfs['horsepower'], bins = 3)
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('horsepower')
```
